```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def avto_reg_der(x, n = 2, alpha = 10):
    return x**(n+1) + x - alpha

def avto_reg_der2(x, n = 2, alpha = 10):
    return (n+1)*(x**(n)) + 1
```

```python
def Div_2(func, a = 0, b = 10, max_iter = 1000, eps = 0.001, history = None):
    for i in range(max_iter):
        c = (a+b)/2
        res_c = func(c)
        if(history != None):
            history[0].append(abs(a-b))
            history[1].append(abs(res_c))
        #if(abs(a-b) < eps):
        if(abs(res_c) < eps):
            return c
        if (np.sign(res_c) == np.sign(func(a))):
            a = c
        else:
            b = c
    return c
```

```python
max_iter = 1000
eps = 10**-6
```

```python
def avto_reg_der_f(x, n = 2, alpha = 10):
    return -(alpha*n*(x**(n-1))/((1 + x**n)**2) + 1)

x = list(np.linspace(0, 10, 10000))
figure, axis = plt.subplots(1, 1)
figure2, axis2 = plt.subplots(1, 1)
axis.set_xlabel("alpha")
axis.set_ylabel("x*")
axis2.set_xlabel("alpha")
axis2.set_ylabel("b")

for n in [2, 4, 6]:
    history = []
    history_b = []
    for alpha in x:
        def avto_reg_der_run(x):
            return avto_reg_der(x, n, alpha)
        
        x0 = Div_2(avto_reg_der_run, a = 0, b = 10, max_iter=max_iter, eps=eps)
        history.append(x0)
        history_b.append(-n/alpha*(x0**(n+1)))
    axis.plot(x, history, label = "n = " + str(n))
    axis2.plot(x, history_b, label = "n = " + str(n))
axis2.plot(x, -1*np.ones(len(x)), label = "b = -1")

figure.tight_layout()
figure.legend()
axis.grid(True)
figure2.tight_layout()
figure2.legend()
axis2.grid(True)
figure.show()
figure2.show()
```
